###  **4. (4 điểm) Tính correlation giữa Hours_Worked và Performance cho từng Location. Dựa trên kết quả, đề xuất một chiến lược phân bổ công việc khác nhau cho từng Location để tối ưu hóa hiệu suất, giải thích tại sao chiến lược này phù hợp với đặc điểm nhân viên tại mỗi địa điểm.**

```python
import pandas as pd

# Tính correlation giữa Hours_Worked và Performance cho từng Location
correlation_by_location = df.groupby('Location').apply(lambda x: x['Hours_Worked'].corr(x['Performance']))
print("Hệ số tương quan giữa Hours_Worked và Performance cho từng Location:")
print(correlation_by_location)

# Đề xuất chiến lược phân bổ công việc
print("\nĐề xuất chiến lược phân bổ công việc:")
for location, corr in correlation_by_location.items():
    if corr > 0.5:
        strategy = "Tăng số giờ làm việc để cải thiện hiệu suất."
    elif corr < -0.5:
        strategy = "Giảm số giờ làm việc để tránh kiệt sức và cải thiện hiệu suất."
    else:
        strategy = "Tập trung vào các yếu tố khác như đào tạo hoặc công cụ hỗ trợ thay vì thay đổi số giờ làm việc."
    print(f"- {location}: {strategy}")
```

## **5. (4 điểm) Vẽ boxplot của Performance theo Department và Gender (kết hợp) bằng Seaborn. Xác định outlier bằng IQR, sau đó đề xuất một kế hoạch phỏng vấn cá nhân với các nhân viên outlier để tìm hiểu nguyên nhân, giải thích cách kế hoạch này cải thiện quản trị nhân sự.**


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Vẽ boxplot của Performance theo Department và Gender
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Department', y='Performance', hue='Gender', palette='Set2')
plt.title('Boxplot của Performance theo Department và Gender', fontsize=14)
plt.xlabel('Department', fontsize=12)
plt.ylabel('Performance', fontsize=12)
plt.legend(title='Gender')
plt.show()

# Xác định outlier bằng IQR
def find_outliers(group):
    Q1 = group['Performance'].quantile(0.25)
    Q3 = group['Performance'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return group[(group['Performance'] < lower_bound) | (group['Performance'] > upper_bound)]

# Áp dụng hàm để tìm outlier theo Department và Gender
outliers = df.groupby(['Department', 'Gender']).apply(find_outliers).reset_index(drop=True)
print("Danh sách nhân viên outlier:")
print(outliers)

# Đề xuất kế hoạch phỏng vấn cá nhân
print("\nKế hoạch phỏng vấn cá nhân với các nhân viên outlier:")
print("1. Chuẩn bị:")
print("   - Thu thập thông tin chi tiết về hiệu suất, công việc, và môi trường làm việc của từng nhân viên outlier.")
print("2. Phỏng vấn:")
print("   - Đối với nhân viên có hiệu suất thấp: Hỏi về khó khăn trong công việc, kỹ năng cần cải thiện, hoặc các yếu tố ảnh hưởng đến hiệu suất.")
print("   - Đối với nhân viên có hiệu suất cao: Hỏi về phương pháp làm việc, công cụ hỗ trợ, hoặc các yếu tố giúp họ đạt hiệu suất cao.")
print("3. Hành động:")
print("   - Đối với nhân viên có hiệu suất thấp: Đề xuất đào tạo, hỗ trợ, hoặc thay đổi môi trường làm việc.")
print("   - Đối với nhân viên có hiệu suất cao: Đề xuất khen thưởng, thăng chức, hoặc chia sẻ kinh nghiệm với đồng nghiệp.")
```

## **7. (4 điểm) Tính tỷ lệ nhân viên nữ (Gender = Female) trong từng Department sau khi điền giá trị thiếu bằng mode. Dựa trên kết quả, đề xuất một chính sách đa dạng giới tính cho công ty, giải thích cách chính sách này ảnh hưởng đến văn hóa tổ chức.**

## **11. (4 điểm) Điền giá trị thiếu trong Performance bằng hồi quy tuyến tính thủ công dựa trên Hours_Worked và Training_Hours. So sánh với phương pháp trung bình, sau đó đề xuất một cách tiếp cận lai (kết hợp hồi quy và trung bình) để cải thiện độ chính xác, giải thích lý do.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Chuẩn bị dữ liệu
df_complete = df.dropna(subset=['Performance', 'Hours_Worked', 'Training_Hours'])  # Dữ liệu không thiếu
X = df_complete[['Hours_Worked', 'Training_Hours']].values  # Biến độc lập
y = df_complete['Performance'].values  # Biến phụ thuộc

# Tính toán tham số hồi quy tuyến tính
X = np.hstack([np.ones((X.shape[0], 1)), X])  # Thêm cột 1 vào X để tính hệ số chặn
beta = np.linalg.inv(X.T @ X) @ X.T @ y  # Công thức hồi quy tuyến tính

# Điền giá trị thiếu bằng hồi quy tuyến tính
df_missing = df[df['Performance'].isnull()]  # Dữ liệu thiếu
X_missing = df_missing[['Hours_Worked', 'Training_Hours']].values
X_missing = np.hstack([np.ones((X_missing.shape[0], 1)), X_missing])  # Thêm cột 1 vào X_missing
df.loc[df['Performance'].isnull(), 'Performance'] = X_missing @ beta  # Điền giá trị thiếu

# Điền giá trị thiếu bằng trung bình
df_mean = df.copy()
df_mean['Performance'] = df_mean['Performance'].fillna(df_mean['Performance'].mean())

# So sánh phân phối dữ liệu
plt.figure(figsize=(10, 6))
plt.hist(df['Performance'], bins=20, alpha=0.5, label='Hồi quy tuyến tính', color='blue')
plt.hist(df_mean['Performance'], bins=20, alpha=0.5, label='Trung bình', color='orange')
plt.title('So sánh phân phối Performance sau khi điền giá trị thiếu', fontsize=14)
plt.xlabel('Performance', fontsize=12)
plt.ylabel('Tần suất', fontsize=12)
plt.legend()
plt.show()

# Cách tiếp cận lai
df_combined = df.copy()

# Điền giá trị thiếu bằng hồi quy tuyến tính nếu có đủ dữ liệu
df_combined.loc[df_combined['Performance'].isnull() & df_combined['Hours_Worked'].notnull() & df_combined['Training_Hours'].notnull(), 'Performance'] = (
    X_missing @ beta
)

# Điền giá trị thiếu còn lại bằng trung bình
df_combined['Performance'] = df_combined['Performance'].fillna(df_combined['Performance'].mean())
```

## **15. (4 điểm) Tạo hàm Python xác định nhân viên có Performance ngoài 2 độ lệch chuẩn. Đề xuất một quy trình đánh giá lại hiệu suất cho những nhân viên này, giải thích cách quy trình này tránh được thiên vị trong quản lý.**

```python
import pandas as pd

def find_outliers(df, column):
    """
    Xác định nhân viên có Performance ngoài 2 độ lệch chuẩn.
    
    Args:
    - df (DataFrame): Dữ liệu nhân viên.
    - column (str): Tên cột cần kiểm tra (ví dụ: 'Performance').
    
    Returns:
    - DataFrame: Danh sách nhân viên có Performance ngoài 2 độ lệch chuẩn.
    """
    mean = df[column].mean()
    std = df[column].std()
    lower_bound = mean - 2 * std
    upper_bound = mean + 2 * std
    
    # Lọc nhân viên có Performance ngoài khoảng [mean - 2*std, mean + 2*std]
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Áp dụng hàm để tìm nhân viên có Performance ngoài 2 độ lệch chuẩn
outliers = find_outliers(df, 'Performance')
print("Nhân viên có Performance ngoài 2 độ lệch chuẩn:")
print(outliers)

# Đề xuất quy trình đánh giá lại hiệu suất
print("\nĐề xuất quy trình đánh giá lại hiệu suất:")
print("1. Phân tích nguyên nhân:")
print("   - Đối với nhân viên có hiệu suất thấp: Xác định các yếu tố như thiếu kỹ năng, thiếu đào tạo, hoặc môi trường làm việc không phù hợp.")
print("   - Đối với nhân viên có hiệu suất cao: Xác định các yếu tố như kỹ năng vượt trội, công cụ hỗ trợ, hoặc cách làm việc hiệu quả.")
print("2. Phỏng vấn trực tiếp:")
print("   - Trao đổi với nhân viên để hiểu rõ hơn về các yếu tố ảnh hưởng đến hiệu suất.")
print("3. Đánh giá lại:")
print("   - Sử dụng các tiêu chí đánh giá công bằng, dựa trên dữ liệu và phản hồi từ nhân viên.")
print("4. Hành động cụ thể:")
print("   - Đối với nhân viên có hiệu suất thấp: Đề xuất đào tạo, hỗ trợ hoặc thay đổi môi trường làm việc.")
print("   - Đối với nhân viên có hiệu suất cao: Đề xuất khen thưởng, thăng chức, hoặc chia sẻ kinh nghiệm với đồng nghiệp.")
```

## **18. (15 điểm) Xây dựng ANN bằng PyTorch với kiến trúc như hình sau. Huấn luyện 20 epochs với CrossEntropyLoss và Adam (lr=0.001), dùng early stopping (patience=5) dựa trên loss của tập test. Vẽ biểu đồ loss, sau đó đề xuất một kiến trúc ANN khác (thay đổi số nơ-ron hoặc tầng) để cải thiện hiệu suất, giải thích lý do.**

```python
import torch.nn as nn

# Xây dựng ANN
class ANN(nn.Module):
    def __init__(self):
        super(ANN, self).__init__()
        self.fc1 = nn.Linear(28*28, 128)  # Tầng ẩn 1
        self.fc2 = nn.Linear(128, 64)    # Tầng ẩn 2
        self.fc3 = nn.Linear(64, 10)     # Tầng đầu ra (10 lớp)
        self.relu = nn.ReLU()            # Hàm kích hoạt ReLU

    def forward(self, x):
        x = x.view(x.size(0), -1)        # Flatten ảnh 28x28 thành vector 784 chiều
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
import torch
from torch.optim import Adam
from torch.utils.data import DataLoader

# Khởi tạo mô hình, hàm mất mát và optimizer
model = ANN()
criterion = nn.CrossEntropyLoss()
optimizer = Adam(model.parameters(), lr=0.001)

# Huấn luyện mô hình
def train_model(model, train_loader, test_loader, criterion, optimizer, epochs=20, patience=5):
    train_loss_history = []
    test_loss_history = []
    best_loss = float('inf')
    patience_counter = 0

    for epoch in range(epochs):
        # Huấn luyện
        model.train()
        train_loss = 0
        for images, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
        train_loss /= len(train_loader)
        train_loss_history.append(train_loss)

        # Đánh giá trên tập test
        model.eval()
        test_loss = 0
        with torch.no_grad():
            for images, labels in test_loader:
                outputs = model(images)
                loss = criterion(outputs, labels)
                test_loss += loss.item()
        test_loss /= len(test_loader)
        test_loss_history.append(test_loss)

        print(f"Epoch {epoch+1}/{epochs}, Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}")

        # Early stopping
        if test_loss < best_loss:
            best_loss = test_loss
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print("Early stopping triggered.")
                break

    return train_loss_history, test_loss_history

# Huấn luyện mô hình
train_loss_history, test_loss_history = train_model(model, train_loader, test_loader, criterion, optimizer)
```

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ loss
plt.figure(figsize=(8, 6))
plt.plot(train_loss_history, label='Train Loss')
plt.plot(test_loss_history, label='Test Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss over Epochs')
plt.legend()
plt.show()
```

## **19. (5 điểm) Đánh giá mô hình bằng accuracy và confusion matrix trên tập test tự chia. Nếu accuracy dưới 95%, phân tích các chữ số bị nhầm lẫn nhiều nhất trong confusion matrix, đề xuất một cách cải thiện mô hình dựa trên đặc điểm của những chữ số này (không chỉ dùng Dropout hay tăng epochs).**

```python
import torch
from sklearn.metrics import confusion_matrix
import numpy as np

# Hàm tính accuracy
def calculate_accuracy(model, test_loader):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in test_loader:
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    accuracy = 100 * correct / total
    return accuracy

# Tính accuracy trên tập test
accuracy = calculate_accuracy(model, test_loader)
print(f"Accuracy trên tập test: {accuracy:.2f}%")

# Tính confusion matrix
def get_confusion_matrix(model, test_loader):
    model.eval()
    all_labels = []
    all_predictions = []
    with torch.no_grad():
        for images, labels in test_loader:
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            all_labels.extend(labels.numpy())
            all_predictions.extend(predicted.numpy())
    cm = confusion_matrix(all_labels, all_predictions)
    return cm

conf_matrix = get_confusion_matrix(model, test_loader)
print("Confusion Matrix:")
print(conf_matrix)

# Phân tích các chữ số bị nhầm lẫn nhiều nhất
def analyze_confusion_matrix(cm):
    max_confusion = np.unravel_index(np.argmax(cm - np.diag(np.diag(cm))), cm.shape)
    print(f"Chữ số bị nhầm lẫn nhiều nhất: {max_confusion[0]} -> {max_confusion[1]}")
    print(f"Số lần nhầm lẫn: {cm[max_confusion]}")

analyze_confusion_matrix(conf_matrix)

# Đề xuất cách cải thiện mô hình
print("\nĐề xuất cách cải thiện mô hình:")
print("1. Tăng cường dữ liệu (Data Augmentation) như xoay, phóng to/thu nhỏ, hoặc làm nhiễu ảnh.")
print("2. Sử dụng kỹ thuật Attention để mô hình tập trung vào các phần quan trọng của ảnh.")
print("3. Thêm Regularization (L2 Regularization) để giảm overfitting.")
print("4. Thay ANN bằng CNN để học các đặc trưng không gian của ảnh.")
```
