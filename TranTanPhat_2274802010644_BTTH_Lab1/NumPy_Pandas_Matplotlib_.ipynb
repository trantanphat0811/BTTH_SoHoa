## 1. NUMPY



1.1 Tạo mảng trong numpy

```python
import numpy as np

#Tạo mảng 1 chiều
a= np.array([1,2,3,4,5])
print(a)

#Tạo mảng 2 chiều
b= np.array([[1,2,3],[4,5,6]])
print(b)

#Tạo mảng ngẫu nhiên
c= np.random.rand(4,4)
print(c)
```

1.2 Các thao tác cơ bản trong Numpy


```python
#Tính cộng và nhân với 10
print(a *10)
print(a +10)

#Tính tổng, trung bình, lớn nhất, nhỏ nhất
print(a.sum())
print(a.mean())
print(a.max())
print(a.min())

#Lấy phần tử 
print(a[3])



```

## 2. Pandas

## 2.1: Tạo DataFrame


```python
import pandas as pd

#Taoj dataFrame vaf dictionary
data= {
    "Tên": ["A","B","C","D"],
    "Tuổi": [20,18,30,22],
    "Điểm": [9,8,7,5]
}

df= pd.DataFrame(data)
print(df)
```

## 2.2 Thao tác cơ bản

```python

print (df["Điểm"])
print (df.iloc[1])

print(df[df["Tuổi"]>=20])

#Tính tuổi trung bình
print(df["Tuổi"].mean())
```

## 3. Matplotlib


3.1 Biểu đồ đường 

```python
import matplotlib.pyplot as plt
x= [1,2,3,4,5]
y=[6,7,8,9,10]

plt.plot(x,y, label= "Biểu đồ đường", color="pink", marker= "o")
plt.title("Biểu đồ đường")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.legend()
plt.show()
```

3.2 Biểu đồ cột (Bar)

```python
Tên = [ "A","B","C","D"]
Điểm= [5,6,8,9]

plt.bar(Tên, Điểm, color= "pink")
plt.title("Biểu đồ cột")
plt.xlabel("Tên")
plt.ylabel("Điểm")
plt.show()
```

3.3 Biểu đồ phân tán (Scatter)

```python
import numpy as np
x= np.random.rand(100)
y= np.random.rand(100)

sizes= np.random.rand(100)*50
plt.scatter(x,y,s=sizes, alpha=0.5,c="pink")
plt.title("Biểu đồ phân tán")
plt.xlabel("Trục x")
plt.ylabel("Trục x")
plt.show()
```
